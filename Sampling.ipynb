# Importing Data

```python
import pandas as pd
data=pd.read_csv("/content/Creditcard_data.csv")
data
```

# Checking accuracy of original Dataset

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,roc_auc_score


y_train=data['Class']
x_train=data.drop(columns='Class')
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size = 0.2, random_state =43 )

X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.30,random_state=40)
xgb_model = XGBClassifier().fit(x_train, y_train)

xgb_y_predict = xgb_model.predict(x_train)

xgb_score = accuracy_score(xgb_y_predict, y_train)

print('Accuracy score is:', xgb_score)
```

# Balancing Dataset

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
ros = RandomOverSampler(random_state=42)

x_ros, y_ros = ros.fit_resample(x_train, y_train)

print('Original dataset shape', Counter(y_train))
print('Resample dataset shape', Counter(y_ros))
```

# Testing for accuracy of new dataset

```python
xgb_model = XGBClassifier().fit(x_ros, y_ros)

xgb_y_predict = xgb_model.predict(x_ros)

xgb_score = roc_auc_score(xgb_y_predict, y_ros)

print('Accuracy score is:', xgb_score)
```

# Balanced Dataset


```python
df=pd.DataFrame(x_ros)
df['Class']=y_ros
df
```

# Simple Random Sampling

```python
import math
z = 1.96
p = 0.5
E = 0.05

ss = math.ceil((z**2 * p * (1 - p)) / (E**2))
s1=df.sample(n=ss,random_state=30)
sample_set=[]
sample_set.append(s1)
s1['Class'].value_counts()
```

# Systematic Sampling

```python
n=len(df)
k=int(math.sqrt(n))
s2=df.iloc[::k]
sample_set.append(s2)
s2['Class'].value_counts()
```

# Stratified Sampling

```python
z = 1.96
p = 0.5
E = 0.05
S=  0.7
ss=round((z**2*p*(1-p))/((E/S)**2))
s3=df.groupby('Class',group_keys=False).apply(lambda x: x.sample(ss))
sample_set.append(s3)
s3['Class'].value_counts()
```

# Cluster Sampling

```python
import numpy as np
z = 1.96
p = 0.5
E = 0.05
C = 1.5
ss=round((z**2*p*(1-p))/((E/C)**2))
clusters=2
df_new=df
N = len(df)
K = int(N/ss)
data = None
for k in range(K):
    sample_k = df_new.sample(ss)
    sample_k["cluster"] = np.repeat(k,len(sample_k))
    df_new = df_new.drop(index = sample_k.index)
    data = pd.concat([data,sample_k],axis = 0)

random_chosen_clusters = np.random.randint(0,K,size = clusters)
s4 = data[data.cluster.isin(random_chosen_clusters)]
s4.drop(['cluster'], axis=1, inplace=True)
sample_set.append(s4)
s4['Class'].value_counts()
```

# Bootstrap Sampling

```python
z=1.96
p=0.5
E=0.05
ss=math.ceil((z**2*p*(1-p))/(E**2))

s5=df.sample(n=ss,random_state=54,replace=True)
sample_set.append(s5)
s5['Class'].value_counts()
```

# Checking Goodness of the Samples



```python
from scipy.stats import shapiro
for i in range(5):
  print('Score of sample ',i+1 ,' is ' , shapiro(sample_set[i]))
```

# Making a Table of Model V/S Sampling Technique

```python
techniques=['Simple-Random','Systematic','Stratified','Cluster','Bootstrap']
ans=pd.DataFrame(columns=techniques, index=['Random Forest','SVM','Logistic Regression','XGBoost','KNN'])
```

# Splitting all the Sample Set into Input and Output and Applying Models

```python
for i in range(5):
  j=0
  inp=sample_set[i].drop('Class',axis=1)
  out=sample_set[i]['Class']
  inp_train, inp_test, out_train, out_test = train_test_split(inp ,out , random_state=53,test_size=0.25, shuffle=True)

  # Applying Random Forest Classifier
  from sklearn.ensemble import RandomForestClassifier
  from sklearn.metrics import accuracy_score
  clf = RandomForestClassifier(n_estimators = 100)
  clf.fit(inp_train, out_train)
  y_pred = clf.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  ans.iloc[j,i]=acc*100


  # Applying SVM
  from sklearn.svm import SVC
  clf = SVC(kernel='rbf')
  clf.fit(inp_train, out_train)
  y_pred=clf.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  ans.iloc[j+1,i]=acc*100


  # Applying Logistic Regression
  from sklearn.linear_model import LogisticRegression
  clf = LogisticRegression(random_state = 0,max_iter=2000)
  clf.fit(inp_train, out_train)
  y_pred = clf.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  ans.iloc[j+2,i]=acc*100


  # Applying XGBoost Classifier
  from xgboost import XGBClassifier
  xgb_model = XGBClassifier().fit(inp_train, out_train)
  xgb_y_predict = xgb_model.predict(inp_train)
  acc = accuracy_score(xgb_y_predict, out_train)
  ans.iloc[j+3,i]=acc*100


  # Applying KNN
  from sklearn.preprocessing import StandardScaler
  from sklearn.neighbors import KNeighborsClassifier
  scaler=StandardScaler()
  inp_train_scaled=scaler.fit_transform(inp_train)
  inp_test_scaled=scaler.fit_transform(inp_test)

  knn=KNeighborsClassifier()
  knn.fit(inp_train_scaled,out_train)

  out_pred=knn.predict(inp_test)
  acc=accuracy_score(out_test,out_pred)
  ans.iloc[j+4,i]=acc*100

print(ans)
```
